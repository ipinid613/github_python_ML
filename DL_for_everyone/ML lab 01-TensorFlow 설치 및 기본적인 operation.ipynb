# Computational Graph

1. Build Graph(tensors) using TensorFlow Operations
2. Feed data and run Graph(operation)
3. Update variables in the Graph(and return values)

````python
#Build Graph(tensors) using TensorFlow Operations

node1 = tf.constant(3.0, tf.float32)
node2 = tf.constant(4.0)
node3 = tf.add(node1, node2) # node3 = node1 + node2
````

````python
print('node1:', node1, 'node2:', node2)
print('node3:', node3)
````

````python
#Feed data and run Graph(operation)
tf.print('node1, node2:', [node1, node2])

#Update variables in the Graph(and return values)
tf.print('node3:', node3)
````

````python
a = tf.placeholder(tf.float32)
b = tf.placeholder(tf.float32)
adder_node = a + b # + provides a shortcut for tf.add(a, b)

tf.print(adder_node, feed_dict={a: 3, b: 4.5})
tf.print(adder_node, feed_dict={a: [1, 3], b: [2, 7]})
````

````python
# Define the function with inputs
@tf.function
def add(a, b):
    return a + b

# Define the function to print the result
def print_result(a_value, b_value):
    a = tf.constant(a_value, dtype=tf.float32)
    b = tf.constant(b_value, dtype=tf.float32)
    result = add(a, b)
    tf.print(result)

# Run the function with different inputs
print_result(3, 4.5)          # Output: 7.5
print_result([1, 3], [2, 7])  # Output: [3, 10]
````

# Tensor Ranks

랭크는 텐서의 차원 수를 나타냅니다. 랭크가 증가할수록 데이터 구조는 더 복잡해집니다.

스칼라 (Scalar): 0차원 텐서 (랭크 0)
```python
s = tf.constant(5)
```
---
벡터 (Vector): 1차원 텐서 (랭크 1)
```python
v = tf.constant([1, 2, 3])
```
---
행렬 (Matrix): 2차원 텐서 (랭크 2)
```python
m = tf.constant([[1, 2], [3, 4]])
```
---
3차원 텐서: 3차원 텐서 (랭크 3)
```python
t3 = tf.constant([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
```
N차원 텐서: N차원 텐서 (랭크 N)

랭크가 증가할수록 텐서의 차원이 더 많아지고, 그만큼 데이터 구조가 복잡해집니다.

# Tensor Shapes

**형태(shape)**는 텐서의 각 차원에서의 크기를 나타냅니다. 형태는 각 차원의 크기를 나타내는 정수들의 튜플입니다.

스칼라 (Scalar): shape = []
```python
s = tf.constant(5)
print(s.shape)  # Output: ()
```
---
벡터 (Vector): shape = [D0]
```python
v = tf.constant([1, 2, 3])
print(v.shape)  # Output: (3,)
```
---
행렬 (Matrix): shape = [D0, D1]
```python
m = tf.constant([[1, 2], [3, 4]])
print(m.shape)  # Output: (2, 2)
```
---
3차원 텐서: shape = [D0, D1, D2]
```python
t3 = tf.constant([[[1, 2], [3, 4]], [[5, 6], [7, 8]]])
print(t3.shape)  # Output: (2, 2, 2)
```

# Tensor Types

**타입(type)**은 텐서에 저장되는 데이터의 타입을 나타냅니다. TensorFlow는 다양한 데이터 타입을 지원합니다. 예를 들어, tf.float32, tf.int32, tf.string 등이 있습니다.

스칼라 텐서의 데이터 타입

```python
s = tf.constant(5, dtype=tf.int32)
print(s.dtype)  # Output: <dtype: 'int32'>
```
---
벡터 텐서의 데이터 타입
```python
v = tf.constant([1.0, 2.0, 3.0], dtype=tf.float32)
print(v.dtype)  # Output: <dtype: 'float32'>
```
---
문자열 텐서의 데이터 타입
```python
str_t = tf.constant(["hello", "world"], dtype=tf.string)
print(str_t.dtype)  # Output: <dtype: 'string'>
```
