# DataFrame 변형하기

## Groupby
- 데이터 분할(split)
- 적용(applying)
- 데이터 병합(combine)

```dataframe.gropuby('성별').mean()``` <- 성별 별로 평균값 구하기

### 1-1. 데이터 읽기

````python
sample = pd.read_csv("Clean_Dataset.csv")
sample
````

### 1-2. 그룹화 하기

````python
airline_group = sample.groupby('airline')
airline_group
````

````python
# 그룹별로 묶였는지 확인: groups
airline_group.groups
````

### 1-3. groupby 내부 함수 활용하기
- count
- size: 집단별 크기
- sum
- mean, std, var
- min, max

````python
airline_group.count()
````

````python
airline_group.size()
````

````python
# 범주형 자료에 문자가 포함될 경우 에러 발생할 수 있음
airline_group.mean()
````

````python
# 숫자 타입의 컬럼들만 선택
numeric_columns = sample.select_dtypes(include='number').columns

# 그룹바이 후 해당 컬럼들에 대해서만 평균 계산
result = sample.groupby('airline')[numeric_columns].mean()
result
````

````python
sample.info()
````

````python
# 아니면 아렇게 숫자형 컬러만 수동으로 찾아서 처리
airline_group[ ['price', 'days_left', 'duration' ] ].mean()
````

````python
airline_group['price'].max()
airline_group.max()['price'] # 똑같음
````

````python
# 위 절차 총정리
sample.groupby('airline')[ ['price', 'days_left', 'duration' ] ].mean()
````

### 1-4. 그룹화 하기(여러 개)

````python
# airline, source_city 별로 그룹화해서 price 평균 구하기
sample.groupby( ['airline', 'source_city'] )['price'].mean()

# 변수화 해서도 가능함
multi_group = sample.groupby( ['airline', 'source_city'] )
multi_group['price'].mean()
````

````python
# 멀티인덱스에서 loc 함수를 사용하여 원하는 row 선택하기

sample.groupby( ['airline', 'source_city'] )['price'].mean().loc[ [('SpiceJet', 'Delhi'), ('Vistara', 'Delhi')] ]
````

## pivot_table
- DataFrame 형태를 변경
- 새로운 인사이트를 발견할 수 있음
- 새로운 데이터 생성 가능하므로 퓨처 엔지니어링에 활용 가능


```
pandas.pivot_table(data, index, columns, aggfunc)
# 예를 들면 이렇게
pd.pivot_table(data=sample, index='고객ID', columns='상품코드', values='구매금액', aggfunc='mean')
```



## stack, unstack

### stack
- 컬럼 레벨로 보이는 컬럼명들을 인덱스 레벨로 가져와 DataFrame 변경

### unstack
- 인덱스 레벨에서 보이는 인덱스값을 컬럼 레벨로 가져와 DataFrame 변경
