````python
multi_sample.head(3)
````

### 1-6. aggregate(집계) 함수 사용하기
- groupby 결과에 집계함수를 적용하여 그룹별 데이터 확인 가능(count, mean, max 등).
- 단, mean 함수 사용 시 문자열 포함으로 인한 오류 안 나게 유의

````python
# aggregate 사용해보기
sample.set_index(['airline', 'source_city']).groupby(['airline', 'source_city']).aggregate(['min', 'max', 'count']).head()
````

````python
# 원하는 컬럼만 지정
sample.set_index(['airline', 'source_city'])[ ['duration', 'days_left', 'price'] ].groupby(['airline', 'source_city']).aggregate(['min', 'max', 'count']).head()
````

## pivot_table
- DataFrame 형태를 변경
- 새로운 인사이트를 발견할 수 있음
- 새로운 데이터 생성 가능하므로 퓨처 엔지니어링에 활용 가능


```
pandas.pivot_table(data, index, columns, aggfunc)
# 예를 들면 이렇게
pd.pivot_table(data=sample, index='고객ID', columns='상품코드', values='구매금액', aggfunc='mean')
```



````python
# 딕셔너리를 이용해서 dataframe 만들기

pivot_sample = pd.DataFrame({
    '고객ID': ['cust_1', 'cust_1', 'cust_1', 'cust_2', 'cust_2', 'cust_2', 'cust_3', 'cust_3', 'cust_3'],
    '상품코드': ['p1', 'p2', 'p3', 'p1', 'p2', 'p3', 'p1', 'p2', 'p3'],
    '등급': ['A', 'A', 'A', 'A', 'A', 'A', 'B', 'B', 'B'],
    '구매금액': [30, 10, 0, 40, 15, 30, 0, 0, 10],
    '나이': [15, 40, 33, 29, 6, 30, 54, 69, 38]
})

pivot_sample
````

````python
pivot_sample.info()
````

### 2-1. pivot_table() 함수

- **index, columns에는 '범주형' 컬럼 지정해야 함!**

````python
# pivot_table() 함수 사용해보기
# aggfunc 옵션의 디폴트는 mean이다. aggfunc 옵션이 없어도 결과는 똑같다.
# 하지만 상품코드와 등급 컬럼은 문자열이어서 평균을 구할 수 없어 에러 발생

pd.pivot_table( data=pivot_sample, index='고객ID', aggfunc='mean' )
````

````python
# 이럴 땐 이렇게 컬럼을 지정해놓고 수행!
pd.pivot_table( data=pivot_sample[ ['고객ID', '구매금액', '나이'] ], index='고객ID', aggfunc='mean' )
````

````python
# 피봇이랑 이렇게 그룹핑 하는거랑 결과는 동일한 것 확인
pivot_sample[['고객ID', '구매금액', '나이']].groupby('고객ID').mean()
````

````python
# 이번에는 columns 옵션도 넣고 실행해보기
pivot_sample.pivot_table(index='고객ID', columns='상품코드', aggfunc='sum')
````

````python
# 이번에는 values 옵션도 넣고 실행해보기
pivot_sample.pivot_table(index='고객ID', columns='상품코드', values='구매금액', aggfunc='mean')
````

### 2-2. pivot_table의 추가 살펴 보기
- pivot_table은 aggregation 함수를 활용하여 처리

````python
# 집계함수 옵션을 sum으로 하여 합계 구하기
pivot_sample.pivot_table(index='등급', columns='상품코드', values='구매금액', aggfunc='sum')
````

````python
# 위와 같은 결과를 얻기 위한 다른 코드
pd.pivot_table(data=pivot_sample, index='등급', columns='상품코드', values='구매금액', aggfunc=np.sum) #aggfunc='sum'도 됨
````

## stack, unstack

### stack
- 컬럼 레벨로 보이는 컬럼명들을 인덱스 레벨로 가져와 DataFrame 변경
- 즉, 데이터를 row 레벨로 쌓아올리는 개념으로 이해하면 쉬움

### unstack
- 인덱스 레벨에서 보이는 인덱스값을 컬럼 레벨로 가져와 DataFrame 변경
- stack의 반대 operation

````python
# 데이터 프레임 만들기

stack_sample = pd.DataFrame({
    '지역': ['서울', '서울', '서울', '경기', '경기', '부산', '서울', '서울', '부산', '경기', '경기', '경기'],
    '요일': ['월요일', '화요일', '수요일', '월요일', '화요일', '월요일', '목요일', '금요일', '화요일', '수요일', '목요일', '금요일'],
    '강수량': [100, 80, 1000, 200, 200, 100, 50, 100, 200, 100, 50, 100],
    '강수확률': [80, 70, 90, 10, 20, 30, 50, 90, 20, 80, 50, 10]
})

stack_sample
````

### 3-1. 실습

````python
# set_index를 통해 지역, 요일 컬럼을 인덱스로 설정

new_stack_sample = stack_sample.set_index(['지역', '요일'])
new_stack_sample
````

````python
# unstack: new_stack_sample의 0번째 레벨의 인덱스 '지역'을 컬럼으로 이동

new_stack_sample.unstack('지역') #new_stack_sample.unstack(0),  new_stack_sample.unstack(level=0)과 같음
````

````python
# unstack: new_stack_sample의 1번째 레벨의 인덱스 '요일'을 컬럼으로 이동

new_stack_sample.unstack('요일') #new_stack_sample.unstack(1),  new_stack_sample.unstack(level=1)과 같음
````

````python
# stack: new_stack_sample의 0번째 레벨의 컬럼 '강수량, 강수확률'을 컬럼으로 이동

new_stack_sample.stack(level=0)
````

## [실습] 데이터프레임 변형하기

````python
#데이터 프레임 만들기(1안)
data = {
    '지점': ['A', 'A', 'B', 'B'],
    '상품': ['P1', 'P2', 'P1', 'P2'],
    'Q1': [112, 134, 156, 178],
    'Q2': [212, 234, 256, 278],
    'Q3': [312, 334, 356, 378],
    'Q4': [412, 434, 456, 478]
}
````

````python
data_df = pd.DataFrame(data)
data_df
````

````python
#데이터 프레임 만들기(2안)

data_df = pd.DataFrame({
    '지점': ['A', 'A', 'B', 'B'],
    '상품': ['P1', 'P2', 'P1', 'P2'],
    'Q1': [112, 134, 156, 178],
    'Q2': [212, 234, 256, 278],
    'Q3': [312, 334, 356, 378],
    'Q4': [412, 434, 456, 478]
})

data_df
````

````python
# '지점' 컬럼을 기준으로 그룹화하고 평균 구하기
# 단, '상품' 컬럼은 제외

data_df.drop('상품', axis=1).groupby('지점').mean()

#또는
data_df.groupby('지점')[['Q1', 'Q2', 'Q3', 'Q4']].mean()
````

````python
# '상품' 컬럼을 기준으로 그룹화 하고 합계를 구하세요
# 단, '지점' 컬럼은 제외

data_df.drop('지점', axis=1).groupby('상품').aggregate('sum')

#또는
data_df.groupby('상품')[['Q1', 'Q2', 'Q3', 'Q4']].sum()
````

````python
# pivot_table 활용해서 인덱스는 '지점', 컬럼은 '상품', values는 'Q1'으로 평균 구하기

data_df.pivot_table(index='지점', columns='상품', values='Q1', aggfunc='mean')
````

````python
# set_index를 통해 '지점' 컬럼을 인덱스로 만들기
# data_df1에 저장

data_df1 = data_df.set_index('지점')
data_df1
````

````python
# reset_index() 하기

data_df1.reset_index()
data_df1
````

````python
# 멀티인덱스 '지점', '상품'을 인덱스로 만들기
# data_df2에 저장

data_df2 = data_df.set_index(['지점', '상품'])
data_df2
````

````python
# data_df2 프레임에 대해 unstack 함수 사용하여 '상품' 인덱스를 컬럼레벨로 올리기
# data_df3에 저장

data_df3 = data_df2.unstack('상품') #또는 unstack(1)
data_df3
````

````python
# data_df3 프레임에 대해 stack 함수 사용하여 '상품' 컬럼을 인덱스 레벨로 내리기

data_df3.stack('상품') #또는 stack(1)
````
