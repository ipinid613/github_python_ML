## Groupby
- 데이터 분할(split)
- 적용(applying)
- 데이터 병합(combine)

```dataframe.gropuby('성별').mean()``` <- 성별 별로 평균값 구하기

### 1-1. 데이터 읽기

````python
sample = pd.read_csv("Clean_Dataset.csv")
sample
````

### 1-2. 그룹화 하기

````python
airline_group = sample.groupby('airline')
airline_group
````

````python
# 그룹별로 묶였는지 확인: groups
airline_group.groups
````

### 1-3. groupby 내부 함수 활용하기
- count
- size: 집단별 크기
- sum
- mean, std, var
- min, max

````python
airline_group.count()
````

````python
airline_group.size()
````

````python
# 범주형 자료에 문자가 포함될 경우 에러 발생할 수 있음
airline_group.mean()
````

````python
# 숫자 타입의 컬럼들만 선택
numeric_columns = sample.select_dtypes(include='number').columns

# 그룹바이 후 해당 컬럼들에 대해서만 평균 계산
result = sample.groupby('airline')[numeric_columns].mean()
result
````

````python
sample.info()
````

````python
# 아니면 아렇게 숫자형 컬러만 수동으로 찾아서 처리
airline_group[ ['price', 'days_left', 'duration' ] ].mean()
````

````python
airline_group['price'].max()
airline_group.max()['price'] # 똑같음
````

````python
# 위 절차 총정리
sample.groupby('airline')[ ['price', 'days_left', 'duration' ] ].mean()
````

### 1-4. 그룹화 하기(여러 개)

````python
# airline, source_city 별로 그룹화해서 price 평균 구하기
sample.groupby( ['airline', 'source_city'] )['price'].mean()

# 변수화 해서도 가능함
multi_group = sample.groupby( ['airline', 'source_city'] )
multi_group['price'].mean()
````

````python
# 멀티인덱스에서 loc 함수를 사용하여 원하는 row 선택하기

sample.groupby( ['airline', 'source_city'] )['price'].mean().loc[ [('SpiceJet', 'Delhi'), ('Vistara', 'Delhi')] ]
````

### 1-5. index를 이용한 groupby
- set_index: column 데이터를 index 레벨로 변경할 때 사용
- reset_index: 0부터 시작하는 새로운 인덱스 생성하여 인덱스 초기화

````python
sample.head()
````

````python
sample.drop('Unnamed: 0', axis=1, inplace=True) #위에서 확인한대로 불필요한 컬럼은 그냥 제거함(강의에 없는 내용이고 내가 판단해서 지우는 것임)
````

````python
# sample에서 arrival_time을 인덱스로 만들어 보기

sample.set_index('arrival_time')
````

````python
# 여러 컬럼을 인덱스로 만들어 보기

multi_sample = sample.set_index(['airline', 'source_city'])
multi_sample
````

````python
# reset_index 활용: 기존의 인덱스를 버리고, 새로운 0번부터 인덱스 지정 (set_index <-> reset_index)
multi_sample.reset_index() #단순 출력만 함. 저장되는건 아님
````

````python
multi_sample.head(3)
````

````python
# multi_sample 데이터프레임에서 airline으로 그룹핑 해보기

multi_sample.groupby('airline').count()['flight']
````

````python
# multi_sample 데이터프레임에서 source_city로 그룹핑 해보기
multi_sample.groupby('source_city').count()['flight']
````

````python
# 아래처럼도 할 수 있음
# set_index를 통해 인덱스를 만들고, 그룹핑해서 보기

sample.set_index(['airline', 'source_city']).groupby(['airline', 'source_city']).count().head()
````

### 1-6. aggregate(집계) 함수 사용하기
- groupby 결과에 집계함수를 적용하여 그룹별 데이터 확인 가능(count, mean, max 등).
- 단, mean 함수 사용 시 문자열 포함으로 인한 오류 안 나게 유의

````python
# aggregate 사용해보기
sample.set_index(['airline', 'source_city']).groupby(['airline', 'source_city']).aggregate(['min', 'max', 'count']).head()
````

````python
# 원하는 컬럼만 지정
sample.set_index(['airline', 'source_city'])[ ['duration', 'days_left', 'price'] ].groupby(['airline', 'source_city']).aggregate(['min', 'max', 'count']).head()
````

## pivot_table
- DataFrame 형태를 변경
- 새로운 인사이트를 발견할 수 있음
- 새로운 데이터 생성 가능하므로 퓨처 엔지니어링에 활용 가능


```
pandas.pivot_table(data, index, columns, aggfunc)
# 예를 들면 이렇게
pd.pivot_table(data=sample, index='고객ID', columns='상품코드', values='구매금액', aggfunc='mean')
```



````python
# 딕셔너리를 이용해서 dataframe 만들기

pivot_sample = pd.DataFrame({
    '고객ID': ['cust_1', 'cust_1', 'cust_1', 'cust_2', 'cust_2', 'cust_2', 'cust_3', 'cust_3', 'cust_3'],
    '상품코드': ['p1', 'p2', 'p3', 'p1', 'p2', 'p3', 'p1', 'p2', 'p3'],
    '등급': ['A', 'A', 'A', 'A', 'A', 'A', 'B', 'B', 'B'],
    '구매금액': [30, 10, 0, 40, 15, 30, 0, 0, 10],
    '나이': [15, 40, 33, 29, 6, 30, 54, 69, 38]
})

pivot_sample
````

````python
pivot_sample.info()
````

### 2-1. pivot_table() 함수

- **index, columns에는 '범주형' 컬럼 지정해야 함!**

````python
# pivot_table() 함수 사용해보기
# aggfunc 옵션의 디폴트는 mean이다. aggfunc 옵션이 없어도 결과는 똑같다.
# 하지만 상품코드와 등급 컬럼은 문자열이어서 평균을 구할 수 없어 에러 발생

pd.pivot_table( data=pivot_sample, index='고객ID', aggfunc='mean' )
````

````python
# 이럴 땐 이렇게 컬럼을 지정해놓고 수행!
pd.pivot_table( data=pivot_sample[ ['고객ID', '구매금액', '나이'] ], index='고객ID', aggfunc='mean' )
````

````python
# 피봇이랑 이렇게 그룹핑 하는거랑 결과는 동일한 것 확인
pivot_sample[['고객ID', '구매금액', '나이']].groupby('고객ID').mean()
````

````python
# 이번에는 columns 옵션도 넣고 실행해보기
pivot_sample.pivot_table(index='고객ID', columns='상품코드', aggfunc='sum')
````

````python
# 이번에는 values 옵션도 넣고 실행해보기
pivot_sample.pivot_table(index='고객ID', columns='상품코드', values='구매금액', aggfunc='mean')
````

### 2-2. pivot_table의 추가 살펴 보기
- pivot_table은 aggregation 함수를 활용하여 처리

````python
# 집계함수 옵션을 sum으로 하여 합계 구하기
pivot_sample.pivot_table(index='등급', columns='상품코드', values='구매금액', aggfunc='sum')
````

````python
# 위와 같은 결과를 얻기 위한 다른 코드
pd.pivot_table(data=pivot_sample, index='등급', columns='상품코드', values='구매금액', aggfunc=np.sum) #aggfunc='sum'도 됨
````

## stack, unstack

### stack
- 컬럼 레벨로 보이는 컬럼명들을 인덱스 레벨로 가져와 DataFrame 변경

### unstack
- 인덱스 레벨에서 보이는 인덱스값을 컬럼 레벨로 가져와 DataFrame 변경
