### 결측치 처리 특징

- 결측치 처리 특징 1: 실무자 견해가 많이 반영됨
    - 같은 데이터도 실무자 의견에 따라 결과가 달라질 수도 있고, 임의대로 결측치 처리 시 분석결과가 매우 틀어질 수 있음
- 결측치 처리 특징 2: 시간이 많이 투자되어야 함
    - 데이터 현실을 반영한 결측치 처리가 되어야 정확한 데이터 분석 가능
- 결측치 처리 특징 3: 결측치 제거가 가장 쉬운 방법
    - 막대한 데이터 손실 가능
- 결측치 처리 특징4: 단순 대치 시
    - 데이터에서 편향(bias)이 생길 수 있음

### 결측치 채우기
- 값 대체하기:
    - 결측치를 평균화 값으로 대체 - 연속형 값일 경우 평균, 중앙값 | 범주형 값일 경우 최빈값으로 대체
- 예측하기:
    - 결측치를 상관관계 등을 예측하여 대체
    - 결측치의 특성이 무작위로 관찰되는 것이 아니라고 가정, 상관관계나 예측모델을 사용해 채움
- Pros/Cons:
    - Pros: 정보 손실 없이 결측치를 빠르게 채울 수 있음
    - Cons: 모든 결측치가 동일한 값을 가질 수 있음 -> 편향 가능
- Methods:
    - fillna(): 지정한 값으로 값 채우기
    - interpolate(): 연속형 값에서 중간값으로 값 채우기 / 측값과 실측값 사이의 결측값을 마치 '그라데이션(gradation)' 기법으로 색깔을 조금씩 변화시켜가면서 부드럽게 채워나가는 방법


#### 결측치 채우기 - fillna()

```python
# fillna() 함수를 통해 null값에 내가 지정한 값 할당 가능

cust = cust.fillna(15)
cust.head() #기존 null값이 모두 15로 대체됨
```

```python
cust.isnull().sum()
```

```python
# fillna() 함수의 Method 옵션
# 데이터의 주변값으로 결측치를 채울 수 있음

cust = cust_fix.copy() #다시 원본으로부터 복사하기
cust.tail()
```

```python
cust = cust.fillna(method='backfill') #뒤에 있는 값으로 채우기 /// bfill 과 동일
cust.tail() # NaN이었던 데이터가 후행데이터의 값을 가지는 것 확인
```

```python
cust = cust_fix.copy()
cust.head()
```

```python
cust = cust.fillna(method='ffill') #앞에 있는 값으로 채우기
cust.head() # NaN이었던 데이터가 전행데이터의 값을 가지는 것 확인
```

#### 결측치 채우기 - replace()

```python
cust = cust_fix.copy()
```

```python
# age 컬럼의 NaN 값을 중앙값(median)으로 채우기

cust['age'].median() #24.0

cust['age'] = cust['age'].replace(np.NaN, cust[ 'age' ].median())
cust.head() # 2행의 age가 NaN이었는데, 중앙값으로 대체됨
```

#### 결측치 채우기 - interpolate()

```python
# 결측치가 존재하는 앞과 뒤의 값을 보고 그 중간의 값으로 대체

cust = cust_fix.copy()
cust = cust.interpolate()
cust.head() # 2행의 age가 NaN이었는데, 1행과 3행의 중간값으로 대체됨
```

### 결측치 제거하기
- Listwise(dropna()):
    - 결측치가 존재하는 전체 행(row) 삭제 / 일부 변수(col) 값이 존재하더라도 row 통째로 삭제
- Pairwise(dropna(how='all')):
    - 모든 변수(col) 값이 결측치인 경우에만 해당 행(row) 삭제 / 남은 결측치는 데이터 분석 시 무시하게 됨
    - 데이터 손실을 최소화 하지만, 어쨋든 결측치를 포함하기 때문에 데이터 분석이나 최적 모델 만드는데 나쁜 영향을 주기도 함

```python
cust = cust_fix.copy()
cust.info()
```

#### 결측치 제거하기 - Listwise

```python
# Listwise(dropna()):
# 결측치가 존재하는 전체 행(row) 삭제 / 일부 변수(col) 값이 존재하더라도 row 통째로 삭제

cust = cust.dropna()
cust.info() #전체 데이터가 8228개로 많이 소실됨
```

#### 결측치 제거하기 - Pairwise

```python
# Pairwise(dropna(how='all')):
# 모든 변수(col) 값이 결측치인 경우에만 해당 행(row) 삭제 / 남은 결측치는 데이터 분석 시 무시하게 됨
# 데이터 손실을 최소화 하지만, 어쨋든 결측치를 포함하기 때문에 데이터 분석이나 최적 모델 만드는데 나쁜 영향을 주기도 함

cust = cust_fix.copy()
cust = cust.dropna(how='all')
cust.info() #한 줄도 제거가 안 된 모습(모든 컬럼 값이 null인 경우 없음)
```

#### 결측치 제거하기 - Subset 파라미터
- 특정 열 안에서만 삭제

```python
cust = cust_fix.copy()
cust.head()
```

```python
# 특정 컬럼에 NaN이 존재할 경우 해당 라인을 삭제(조건 설정)

cust = cust.dropna(subset=['class'])
cust.head() #1번 행 삭제된 모습
```
