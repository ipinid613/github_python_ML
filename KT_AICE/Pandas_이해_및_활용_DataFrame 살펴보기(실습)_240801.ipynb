## Boolean Indexing으로 Row 선택하기
- 해당 조건에 맞는 row만 선택
- 2개 이상의 조건 묶을 때 AND나 OR을 사용할 수 있음 / &, | 기호로만 사용함

> e.g. source_city가 Delhi 이면서 Duration이 5 이상 8 미만인 항공편만 가져오기



```python
# 조건 분해
# 1. soource_city = Delhi
# 2. Duration 5 이상 7 이하
```

```python
# 1. soource_city = Delhi
a3["source_city"].head()
```

```python
a3["source_city"] == "Delhi"
```

```python
a3 [ a3["source_city"] == "Delhi" ]
```

```python
# 2. Duration 평균 5 이상 7 이하
a3["duration"].head()
```

```python
a3[ (a3["duration"] >= 5) & (a3["duration"] <= 7) ]
```

```python
# 최종: source_city가 Delhi 이면서 Duration이 5 이상 7 이하인 항공편만 가져오기
a3[ (a3["source_city"] == "Delhi") & (a3["duration"] >= 5) & (a3["duration"] <= 7) ]
```

```python
# 조건문이 길어지면 변수화 해서 사용 가능
source_city = (a3["source_city"] == "Delhi")
duration = (a3["duration"] >= 5) & (a3["duration"] <= 7)

a3[ source_city & duration ]
```
