- row, column 동시에 선택하기
    - loc, iloc 속성을 이용할 때, 콤마를 이용하여 row와 col 다 명시 가능

````python
# 102 인덱스의 price 컬럼값 가져오기
a3.loc[ 102, 'price' ]
````

````python
# loc 함수를 활용해 102, 202, 302 인덱스의 airline, source_city, destination_city, price 가져오기
a3.loc[ [102, 202, 302], ['airline', 'source_city', 'destination_city', 'price'] ]
````

````python
# iloc 함수를 활용해 2, 102, 202 인덱스의 airline, source_city, destination_city, price 가져오기
# iloc 함수의 경우 인덱스와 컬럼들의 절대적 위치로 표현하는걸 알 수 있다.

a3.iloc[ [2, 102, 202], [0, 3, 4, 11] ]
````

## Boolean Indexing으로 Row 선택하기
- 해당 조건에 맞는 row만 선택
- 2개 이상의 조건 묶을 때 AND나 OR을 사용할 수 있음 / &, | 기호로만 사용함

> e.g. source_city가 Delhi 이면서 Duration이 5 이상 8 미만인 항공편만 가져오기



````python
# 조건 분해
# 1. soource_city = Delhi
# 2. Duration 5 이상 7 이하
````

````python
# 1. soource_city = Delhi
a3["source_city"].head()
````

````python
a3["source_city"] == "Delhi"
````

````python
a3 [ a3["source_city"] == "Delhi" ]
````

````python
# 2. Duration 평균 5 이상 7 이하
a3["duration"].head()
````

````python
a3[ (a3["duration"] >= 5) & (a3["duration"] <= 7) ]
````

````python
# 최종: source_city가 Delhi 이면서 Duration이 5 이상 7 이하인 항공편만 가져오기
a3[ (a3["source_city"] == "Delhi") & (a3["duration"] >= 5) & (a3["duration"] <= 7) ]
````

````python
# 조건문이 길어지면 변수화 해서 사용 가능
source_city = (a3["source_city"] == "Delhi")
duration = (a3["duration"] >= 5) & (a3["duration"] <= 7)

a3[ source_city & duration ]
````

## 데이터 추가/삭제하기
- 새 Column 추가하기: insert 함수

````python
# price를 두 배로 만들기
a3['price2'] = a3['price'] * 2
a3.head()
````

````python
# 기존 col을 연산해서 새로운 데이터 생성
a3['price3'] = a3['price'] + a3['price2']
a3.head()
````

- Column 삭제하기: drop 함수
    - axis: dataframe에는 차원이 0과 1로 존재함 (0은 행, 1은 열)
    - inplace: True/False : 원본 dataframe 업데이트 여부 결정

````python
# price3 컬럼을 삭제
a3.drop('price3', axis=1) # drop 함수 실행 시 a3 데이터프레임에 저장되는게 아니라, 단순 출력됨 주의 / 원본이 바뀌는게 아님!!
````

````python
# 그래서 다시 출력하면 price3 col 그대로 살아있음
a3.head()
````

````python
# 원본 자체를 지우고 싶다면...
# 방법1: 새로운 데이터프레임에 저장하고 출력
a3_new = a3.drop('price3', axis=1)
a3_new.head()
````

````python
# 방법2: inplace 파라미터를 활용, True인 경우 원본 데이터 바로 Drop
a3.drop('price3', axis=1, inplace=True)
a3.head()
````

- Row 삭제하기: drop 함수
    - axis: dataframe에는 차원이 0과 1로 존재함 (0은 행, 1은 열)
    - inplace: True/False : 원본 dataframe 업데이트 여부 결정

````python
# a3 데이터프레임에서 100, 101, 102행 삭제

a3.drop([100, 101, 102], axis=0)
````
