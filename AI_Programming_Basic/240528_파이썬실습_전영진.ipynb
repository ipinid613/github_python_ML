## 콤마(,)와 에스터리스크(*)를 이용한 import 구문

### 콤마(,)
- 모듈을 import 할 때 콤마(,)를 이용해서 기능을 나열할 수 있음
```python
from calculator import addition
from calculator import subtraction
# -->
from calculator import addition, subtraction
```

### 에스터리스크(*)
- 모듈의 모든 기능을 가져올 때는 에스터리스크(*)를 이용함
```python
from calculator import addition
from calculator import subtraction
from calculator import multiplication
from calculator import division
from calculator import rest
from calculator import portion
# -->
from calculator import *
```

## 자주 사용하는 모듈

````python
import random

print('random : ', random.random()) # 0이상 1미만 난수
print('randint : ', random.randint(0, 9)) # 0이상 9이하 난수
print('randrange : ', random.randrange(0, 9)) # 0이상 9미만 난수
print('sample : ', random.sample(range(0, 10), 3)) # 0이상 10미만 3개 난수

listVar = [1, 2, 3, 4, 5, 6]

# 무작위로 1개 아이템 선택
print('random_choice1 : ', random.choice(listVar))
print('random_choice2 : ', random.choice(listVar))

# 아이템 순서 섞음
random.shuffle(listVar)
print('shuffle : ', listVar)
random.shuffle(listVar)
print('shuffle : ', listVar)
````

````python
import time

lt = time.localtime() # 시스템 시간
print('localTime : ', lt)

# 상수를 이용한 시스템 시간 출력
print('tm_year : ', lt.tm_year) # 년
print('tm_mon : ', lt.tm_mon) # 월
print('tm_mday : ', lt.tm_mday) # 일
print('tm_hour : ', lt.tm_hour) # 시
print('tm_min : ', lt.tm_min) # 분
print('tm_sec : ', lt.tm_sec) # 초
print('tm_wday : ', lt.tm_wday) # 요일
````
