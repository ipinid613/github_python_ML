# 리스트
- 여러 개의 데이터를 하나의 이름으로 저장하는 가장 일반적인 형태
- 각각의 데이터를 항목 또는 원소라고 함
- 대괄호([ ]) 사용

image.png

```python
# 리스트 사용 안 한 예시
b1 = 234
b2 = 82
b3 = 128
b4 = 50
b5 = 155

total = b1 + b2 + b3 + b4 + b5
print("판매 수량 합계 =", total)
print("판매 수량 평균 =", total / 5)
```

```python
# 리스트 사용한 예시
books = [234, 82, 128, 50, 155]
total = 0

for x in books:
    total += x
print("판매 수량 합계 =", total)
print("판매 수량 평균 =", total / len(books))
```

## 리스트 항목의 추가와 삭제

```python
alist = []
print(alist)

print('-' * 20)

alist.append(30)
alist.append('홍길동')
print(alist)

print('-' * 20)

alist.insert(1, '김유신')
print(alist.index('김유신'))
```

```python
alist.remove('홍길동')
print(alist)

del(alist[0])
alist
```

```python
deleted = alist.pop()
print(alist)
print(deleted) # 삭제한 항목이 저장된 변수
```

```python
alist = [1, 2, 3, 4, 5]
alist.clear()
print(alist)
```

## 리스트 정렬

```python
contact = ['정약용', '이순신', '김유신', '유관순']
contact.sort()
print(contact)

contact.sort(reverse = True)
print(contact)
```

```python
numbers = [78, 45, 90, 88, 62]
alist = sorted(numbers)
print(numbers) #원본 리스트
print(alist)   #정렬된 결과를 저장한 리스트
```

```python
numbers = []

for _ in range(5):
    numbers.append(int(input('점수 입력 : ')))

print('평균 = ', sum(numbers) / len(numbers))
```

```python
book = ''
bookList = []
number = 0

print('입력을 종료하려면 [Enter] 키를 누르세요.')
print('=' * 30)

while True:
    book = input('도서명 입력 :')
    if book == '':
        break
    bookList.append(book)

bookList.sort()

print('=' * 30)
for b in bookList:
    number += 1
    print(number, ':', b)
```

```python
menu = 0
bucket = []

while True:
    menu = int(input("메뉴 선택(1. 추가  2. 삭제  0. 종료): "))
    if menu == 1:
        bucket.append(input('추가할 내용: '))
    elif menu == 2:
        bucket.remove(input('삭제할 내용: '))
    elif menu == 0:
        print('프로그램을 종료합니다.')
        break
    else:
        print('메뉴 선택 오류입니다. 다시 선택하세요.')
        continue
    print('*' * 30)
    for x in bucket:
        print(x)
    print('*' * 30)
```
