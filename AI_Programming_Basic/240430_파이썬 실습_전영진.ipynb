# Chapter 09. 튜플과 딕셔너리

- 리스트: 수정, 변경, 추가 모두 가능
- 튜플: 읽기 전용임. 데이터를 load하는 시간이 리스트에 비해서 훨씬 빠름(양이 많을 수록) - 대용량 데이터 읽기용으로 쓰기 좋음 / 편집 불가
    - 데이터가 수정되면 안 되는 예: 회사 급여 명세서
- 딕셔너리: 사전(key-value)

```python
fruits = ('사과', '포도', '수박', '참외', '배', '자두', '복숭아', '바나나')
print(fruits[3]) #참외
print(fruits[-1]) #바나나
```

```python
sports = ('태권도', '야구', '농구', '축구', '배구', '권투', '양궁')

for index, item in enumerate(sports): # 꼭 index, item이라는 이름 명명하지 않아도 됨
    if index % 2 == 1:
        print(index, ' : ', item)
```

```python
midTermResult = {'C/C++': 'A', 'Java': 'B+', 'Mobile': 'A+', 'Hacking': 'F', 'System': 'C+'}

#2 'Java와 시스템 과목의 성적 조회
print('Java와 시스템 과목의 성적 조회')
print(midTermResult['Java'])
print(midTermResult['System'])

print('-' * 30)
#3 추가로 2과목의 성적(파이썬 A, OS A+) 삽입
print('추가로 2과목의 성적(파이썬 A, OS A+) 삽입')
midTermResult['Python'] = 'A'
midTermResult['OS'] = 'A+'

print('-' * 30)
#4 Java와 시스템 성적을 B와 A로 수정
print('Java와 시스템 성적을 B와 A로 수정')
midTermResult['Java'] = 'B'
midTermResult['System'] = 'A'

print('-' * 30)
#5 전체 과목과 성적을 조회하여 최종 성적표 출력
print('전체 과목과 성적을 조회하여 최종 성적표 출력')
for key in midTermResult:
    print(key, '\t\t: ', midTermResult[key])
```

```python
quiz = (['3+2', 5, 3], ['5/2의 몫', 2, 5], ['10-2', 8, 3], ['10^2*2', 200, 5], ['1-(10/4의 나머지)', -1, 5], ['2^4', 16, 3], ['4/2', 2, 3])

answerCount = 0
wrongAnswerCount = 0
totalScore = 0

for item in quiz:
    print('문제 : ', item[0])
    answer = int(input('정답를 입력하세요: '))

    if answer == item[1]:
        answerCount += 1
        totalScore += item[2]
    else:
        wrongAnswerCount += 1

print('-' * 30)
print('정답 개수\t: ', answerCount)
print('오답 개수\t: ', wrongAnswerCount)
print('Total Score\t: ', totalScore)
print('-' * 30)
```
