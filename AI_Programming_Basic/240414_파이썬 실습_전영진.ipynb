# 반복문

```python
x = 1

while x <100:
    print(x, end = " ") # x의 값을 출력하고, 줄 바꿈 대신 한 칸 띄움
    x = x + 2
```

```python
import random

answer = random.randint(1, 20)
number = 0

while answer != number:
    number = int(input('숫자 입력(1~20) : '))
    if number < answer:
        print('위')
    elif number > answer:
        print('아래')
    else:
        print('정답')
```

```python
x = 0

while True: # 무한 반복문
    x += 1
    print(x, end=" ")
```

```python
# 20보다 작은 자연수 중에서 3의 배수만 제외하고 출력
# 반복문을 제어하는 break, continue 사용

x = 0
num = int(input('제외할 배수 입력 : '))

while True: # 무한 반복문. 종료 지점을 잘 정해야 함
    x += 1
    if x >= 20:
        break # while문을 종료
    if x % num == 0:
        continue # while문의 조건식 검사로 이동
    print(x, end=" ")
```

```python
menu = 0      # 선택 메뉴
qunatity = 0    # 주문 수량
price = 0     # 상품 단가
total = 0     # 금액 합계

print("*" * 30)
print("[1]팝콘  [2]나초  [3]핫도그  [4]음료")
print("주문을 끝내려면 [0]을 입력하세요.")
print("-" * 30)

while True:
    menu = int(input("주문할 메뉴를 선택하세요 : "))
    if menu == 0:
        break
    if menu < 1 or menu > 4:
        print("메뉴 선택 오류... 다시 선택하세요 \n")
        continue

    if menu == 1: price = 5000
    elif menu == 2: price = 4000
    elif menu == 3: price = 3500
    else: price = 2000

    qunatity = int(input("수량을 입력하세요 : "))
    total = total + (qunatity * price)
    print()

    print("-" * 30)
    print("금액 합계는", total, "원입니다.")
```

```python
start = int(input("시작할 index : "))
stop = int(input("멈출 index : ")) # 항상 0부터 index 카운팅 됨
step = int(input("증감할 간격 : "))

for i in range(start, stop, step):
    print(i)

### 1, 10, 2 == 1, 3, 5, 7, 9 출력
### 9, 1, -2 == 9, 7, 5, 3, 1 출력
```

```python
s = 0

for x in range(0, 11):
    s += x
    print(s, end=" ")
```

```python
x = 0
s = 0

while x <= 10:
    s = s + x
    print(s, end=" ")
    x = x + 1
```

```python
dan = int(input("몇 단을 출력할까요? : "))

for i in range(1, 10):
    print(f"{dan} x {i} = {dan*i}")

print("-" * 30)
for i in range(1, 10):
    print("%d x %d = %2d" %(dan, i, dan * i))
```
