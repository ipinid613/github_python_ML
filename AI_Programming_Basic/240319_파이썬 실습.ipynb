```python
# ☆핵심: 연산을 해야 하므로 데이터를 int로 형변환 함.
## 형변환을 하지 않으면 str 데이터가 그냥 쭉 나열될 것임.
sales1 = int(input('1월 매출 : '))
sales2 = int(input('2월 매출 : '))
sales3 = int(input('3월 매출 : '))
total = sales1 + sales2 + sales3
print('1분기 전체 매출 : ', total, '원')
```

```python
num1 = 3.54
num2 = 0.25
result = num1 + num2
result

float(result)

int(result)
# >> 3.54 + 0.25 해도 3이 나옴

round(result)
# >> 올림처리 해서 4

round(result, 1)
# >> 1번째 자리까지 반올림
```

```python
# ☆☆☆문자열 연산: 곱하기, 더하기만 가능함!!
print('Apple' * 3)
print('Apple' + 'Apple')

## 단, 0이나 음의 정수를 곱할 때에는 빈 값('') 출력
## ☆☆☆효과: 여러 데이터에서 음수값을 걸러낼 때 이 연산을 많이 사용함 / if문에서 사용하고, 이 연산을 통해 ''(빈 값)을 얻어내고, if('') {실행문} 식으로 많이 씀
print('a' * 0)
# >> ''
print('Hello' * -3)
# >> ''
```

```python
# 신체질량지수(BMI) 구하기

height = float(input('신장(m) : '))
weight = float(input('몸무게(kg) : '))
bmi = int(weight / (height * height))
print('BMI : ', bmi)
```

```python
# 빵을 나누어 줄 수 있는 학생 수 구하기

bread = 97
count = 3
maxCount = bread // count
print('빵을 나누어 줄 수 있는 학생 수 :', maxCount)
rest = bread % count
print('남는 빵 개수 :', rest)
```
