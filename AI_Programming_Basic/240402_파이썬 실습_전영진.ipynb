```python
number = int(input('양의 정수 입력 : '))
result = number / 3
print('최초 나누기 연산', result) # 나누기 연산이므로 float으로 형변환

if (result - int(result)) >= 0.5: # 뺄셈한 값이 0.5 이상이 되면 올림 처리, 그게 아니면 버림 / int로 감싸면 버림처리하게 됨
    result = int(result) + 1
else:
    result = int(result)

print('결과:', result)
```

```python
# ELIF 예제
score = int(input('점수를 입력하세요.'))

if score >= 90:
    print('A')
elif score >= 80 :
    print('B')
elif score >= 70 :
    print('C')
elif score >= 60 :
    print('D')
else:
    print('F')
```

```python
# 잘못 설계된 예제

score = int(input('점수를 입력하세요.'))

if score >= 60:
    print('D')
elif score >= 80 :
    print('B')
elif score >= 70 :
    print('C')
elif score >= 90 :
    print('A')
else:
    print('F')
```

```python
# ELIF + 복합연산 예제

score = int(input('점수를 입력하세요.'))

if score >= 60 and score < 70:
    print('D')
elif score >= 80 and score < 90:
    print('B')
elif score >= 70 and score < 80:
    print('C')
elif score >= 90 :
    print('A')
else:
    print('F')
```
